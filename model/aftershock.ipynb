```python
import pandas as pd
df=pd.read_csv("quake.csv")
df
```

```python
df=df.drop([ 'Type', 'Depth Error','Depth Seismic Stations', 'Magnitude Type',
       'Magnitude Error', 'Magnitude Seismic Stations', 'Azimuthal Gap',
       'Horizontal Distance', 'Horizontal Error', 'Root Mean Square', 'ID',
       'Source', 'Location Source', 'Magnitude Source', 'Status'],axis='columns')
df
```

```python
# Keep only the first timestamp before space
df['Date'] = df['Date'].apply(lambda x: str(x).split()[0])
df['Time'] = df['Time'].apply(lambda x: str(x).split()[0])

# Combine cleaned Date & Time
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], errors='coerce', infer_datetime_format=True)

# Check for invalid rows
df['Year'] = df['Datetime'].dt.year
df['Month'] = df['Datetime'].dt.month
df['Day'] = df['Datetime'].dt.day
df['Hour'] = df['Datetime'].dt.hour
df['Minute'] = df['Datetime'].dt.minute
df['Second'] = df['Datetime'].dt.second

df
```

```python
df['Latitude'] = df['Latitude'].round(2)
df['Longitude'] = df['Longitude'].round(2)
```

```python
df['Magnitude_Category'] = pd.cut(df['Magnitude'],bins=[0,6,7,10],labels=['Moderate','Strong','Major'])
df['Depth_Category'] = pd.cut(df['Depth'],bins=[0,70,300,700],labels=['Shallow','Intermediate','Deep'])
df
```

```python
df['Magnitude_Category'].unique()
```

```python
df['Depth_Category'].unique()
```

```python
df = df.sort_values('Datetime')
df['InterEvent_Time'] = df['Datetime'].diff().dt.total_seconds().fillna(0)
df['Cumulative_Count'] = range(1, len(df)+1)
```

```python
df['Rolling_Mag_Mean'] = df['Magnitude'].rolling(window=10).mean()
df['Events_per_Day'] = df.groupby(df['Datetime'].dt.date)['Magnitude'].transform('count')
df
```

```python
df=df.dropna()
df=df.drop(['Date','Time','Magnitude_Category','Depth_Category','Cumulative_Count','Rolling_Mag_Mean','Events_per_Day'],axis='columns')
df
```

```python
# Example labeling rule (simplified)
df['IsAftershock'] = ((df['Magnitude'] < 6.0) & (df['InterEvent_Time'] < 7*24*3600)).astype(int)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

#scaler = StandardScaler()
y = df['IsAftershock']
X = df.drop(['IsAftershock','Datetime'],axis='columns')
#X_scaled = scaler.fit_transform(X)
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

# Model
clf = RandomForestClassifier(n_estimators=200, random_state=42)
clf.fit(X_train, y_train)

# Prediction


```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
clf.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

y_reg = df['Magnitude']
X_reg = df.drop(['Magnitude', 'Datetime', 'IsAftershock'], axis=1)

clf_magnitude = RandomForestRegressor(n_estimators=200, random_state=42)
clf_magnitude.fit(X_reg, y_reg)
```

```python
import numpy as np
def prediction(lat,long,dep,mag,time,date):
    #loc_index=np.where(X.columns==location.lower())[0][0]
    x=np.zeros(len(X.columns))
    dt = pd.to_datetime(date + " " + time, format="%m/%d/%Y %H:%M:%S")
    last_dt = df["Datetime"].iloc[-1]
    inter_event = (dt - last_dt).total_seconds()
    
    x[0]=lat
    x[1]=long
    x[2]=dep
    x[3]=mag
    x[4]=dt.year
    x[5]=dt.month
    x[6]=dt.day
    x[7]=dt.hour
    x[8]=dt.minute
    x[9]=dt.second
    x[10]=inter_event
    is_aftershock = clf.predict([x])[0]

    if is_aftershock == 1:
        # Step 2: predict magnitude
        x_reg = np.delete(x, X.columns.get_loc("Magnitude"))
        predicted_mag = clf_magnitude.predict([x_reg])[0]
        return {"Aftershock": True, "Predicted_Magnitude": predicted_mag}
    else:
        return {"Aftershock": False, "Predicted_Magnitude": None}
```
